# Visualizing the dataset

The goal of this jupyter notebook is to illustrate how to load the train and validation sets for the MR reconstruction challenge. Note that k-space is not centered.

```python
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
import os
import glob
```

```python
# Train Set 
kspace_files = np.asarray(glob.glob("/Users/robertosouza/GoogleDrive/MR-Data/Normative-Raw-Data/Train/*.npy"))
indexes = np.arange(kspace_files.size,dtype = int)
np.random.shuffle(indexes)
kspace_files_train = kspace_files[indexes]

# Validation set
kspace_files_val = np.asarray(glob.glob("/Users/robertosouza/GoogleDrive/MR-Data/Normative-Raw-Data/Val/*.npy"))


print(kspace_files_train[0])
print(len(kspace_files_train))
print 
print(kspace_files_val[0])
print(len(kspace_files_val))
```

## Sample Train k-space and image reconstruction

```python
sample_kspace_train = np.load(kspace_files_train[0])
sample_rec_train = np.abs(np.fft.ifft2(sample_kspace_train[:,:,:,0]+1j*sample_kspace_train[:,:,:,1]))
slice_index = 90
plt.figure()
plt.subplot(121)
plt.imshow(np.log(1+np.abs(sample_kspace_train[slice_index,:,:,0]+1j*sample_kspace_train[slice_index,:,:,1]))\
           ,cmap = 'gray')
plt.axis("off")
plt.subplot(122)
plt.imshow(sample_rec_train[slice_index],cmap = 'gray')
plt.axis("off")
plt.show()
```

## Sample Validation k-space and image reconstruction

```python
sample_kspace_val = np.load(kspace_files_val[0])
sample_rec_val = np.abs(np.fft.ifft2(sample_kspace_val[:,:,:,0]+1j*sample_kspace_val[:,:,:,1]))
slice_index = 90
plt.figure()
plt.subplot(121)
plt.imshow(np.log(1+np.abs(sample_kspace_val[slice_index,:,:,0]+1j*sample_kspace_val[slice_index,:,:,1]))\
           ,cmap = 'gray')
plt.axis("off")
plt.subplot(122)
plt.imshow(sample_rec_val[slice_index],cmap = 'gray')
plt.axis("off")
plt.show()
```
