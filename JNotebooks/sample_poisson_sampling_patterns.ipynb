```python
%matplotlib inline
import numpy as np
import matplotlib.pylab as plt
import glob
import natsort
```

```python
patterns = glob.glob("../Data/poisson_sampling/*.npy")
patterns = natsort.natsorted(patterns)
print(patterns)
print(len(patterns))
```

```python
plt.figure(figsize=(9,12),dpi = 300)
rows,cols = 5,3

for (ii,p) in enumerate(patterns):
    p = np.load(p)
    R = 1/(p[:,:,:145].sum()/p[:,:,:145].size)
    
    plt.subplot(rows,cols,ii+1)
    plt.imshow(p[0],cmap = "gray")
    plt.axis("off")
    plt.title("R=%.1f" %R)
plt.show()
```
