# Boxplot

This script plots the ordered metrics' boxplots of the challenge. The ordering is from left to right starting with the worse metric and going up all the way to the best metric.

```python
%matplotlib inline
import matplotlib.pylab as plt
import pygal
import numpy as np
import os
import glob
import pickle
import h5py
import sys
import mpld3
```

```python
results = glob.glob("../../../Results-submissions/Metrics/*.pickle")
nslices = 156
nfiles = 50

print("Track 01 - 12-channel")
print("R = 5")
t1_r5 = 'track01_12_channel_R=5'
plot_title_prefix = "Track 01 - R = 5 - "

metrics = ["SSIM", "pSNR", "VIF"]



for (jj,m) in enumerate(metrics):
    data_to_plot = []
    data_mean = []
    labels = []
    for ii in results:
        name = ii.split("/")[-1].split(".pickle")[0]
        res_file = open(ii,'rb')
        res = pickle.load(res_file)
        res_file.close()
    
        if t1_r5 in res:
            res = res[t1_r5]
            res = res[:,jj].reshape(nslices,nfiles)
            labels.append(name)
            data_to_plot.append(res.ravel())
            data_mean.append(res.mean())
    indexes = np.argsort(np.array(data_mean))
    labels = [labels[i] for i in indexes]
    labels = [f.split("-base")[0] for f in labels]
    data_to_plot = [data_to_plot[i] for i in indexes]
    fig = plt.figure(figsize = (18,6))
    plt.boxplot(data_to_plot)
    plt.xticks(range(1,1+len(data_to_plot)), labels, fontweight = "bold")#rotation='vertical',
    plt.ylabel(m)
    plt.title(plot_title_prefix + m )
    mpld3.save_html(fig,(plot_title_prefix + m).replace(" ","") + ".html", template_type='simple')
```
