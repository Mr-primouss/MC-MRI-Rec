# Zero-fille Reconstruction Baseline

This notebook computes the zero-filled reconstruction baseline and saves the results in the same format as the challenge submission structure.

```python
%matplotlib inline
import numpy as np
import matplotlib.pylab as plt
import os
import glob
import h5py
import zipfile
from zero_filled_rec import zero_filled_reconstruction
```

```python
# Make directory for zero-filled reconstruction baseline, Track01 and Track02
results_folder = "zero-filled-baseline"
track01_folder = "zero-filled-baseline/Track01"
track02_folder = "zero-filled-baseline/Track02"

folders = [results_folder,track01_folder,track02_folder]

for ii in folders:
    try:
        os.mkdir(ii)
    except:
        pass
```

## Track 01

### R = 5

```python
# Create results folder
t1_r5_path = "zero-filled-baseline/Track01/12-channel-R=5"
try:
    os.mkdir(t1_r5_path)
except:
    pass
    

# List test set files    
test_track1_r5 = glob.glob("/media/roberto/ROBERTO-5TB/GDrive/CC359/Raw-data/Multi-channel/12-channel/Test-R=5/*.h5")
print("Number of files:",len(test_track1_r5))
print(test_track1_r5[0])

# Plot central slice for each reconstruction
plt.figure(dpi = 300)
cols = 5
rows = 10
slice_index = 120

for (jj,ii) in enumerate(test_track1_r5):
    # Load data
    with h5py.File(ii, 'r') as f:
        zf_kspace = f['kspace']
        rec = zero_filled_reconstruction(zf_kspace) # reconstruct
    
    plt.subplot(rows,cols,jj+1)
    plt.imshow(rec[slice_index],cmap = "gray")
    plt.axis("off")
    
    name  = ii.split("/")[-1] # file name
    
    # Save zero-filled reconstruction
    hf = h5py.File(os.path.join(t1_r5_path,name), 'w')
    hf.create_dataset('reconstruction', data=rec)
    hf.close()

plt.show()
```

## Track 02

### 32-channel

### R = 5

```python
# Create results folder
t2_r5_32c_path = "zero-filled-baseline/Track02/32-channel-R=5"
try:
    os.mkdir(t2_r5_32c_path)
except:
    pass
    

# List test set files    
test_track2_r5_32c = glob.glob("/media/roberto/ROBERTO-5TB/GDrive/CC359/Raw-data/Multi-channel/32-channel/Test-R=5/*.h5")
print("Number of files:",len(test_track2_r5_32c))
print(test_track2_r5_32c[0])

# Plot central slice for each reconstruction
plt.figure(dpi = 300)
cols = 5
rows = 10
slice_index = 120

for (jj,ii) in enumerate(test_track2_r5_32c):
    # Load data
    with h5py.File(ii, 'r') as f:
        zf_kspace = f['kspace']
        rec01 = zero_filled_reconstruction(zf_kspace[:128]) # reconstruct
        rec02 = zero_filled_reconstruction(zf_kspace[128:]) # reconstruct
        rec = np.concatenate((rec01,rec02),axis = 0)
    plt.subplot(rows,cols,jj+1)
    plt.imshow(rec[slice_index],cmap = "gray")
    plt.axis("off")
    
    name  = ii.split("/")[-1] # file name
    
    # Save zero-filled reconstruction
    hf = h5py.File(os.path.join(t2_r5_32c_path,name), 'w')
    hf.create_dataset('reconstruction', data=rec)
    hf.close()

plt.show()
```

```python
# Zip results folder
def zipdir(path, ziph):
    # ziph is zipfile handle
    for root, dirs, files in os.walk(path):
        for file in files:
            ziph.write(os.path.join(root, file))

zipf = zipfile.ZipFile(results_folder + ".zip", 'w', zipfile.ZIP_DEFLATED)
zipdir(results_folder, zipf)
zipf.close()
```
